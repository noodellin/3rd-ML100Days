    主要驗證 Lr 對於grandient 收斂的速度
    

```python
import pandas as pd
```

```python
#cur_x = 3 # The algorithm starts at x=3
precision = 0.000001 #This tells us when to stop the algorithm
max_iters = 10000 # maximum number of iterations
df = lambda x: 2*(x+5) #Gradient of our function 
lrs = [0.1,0.01,0.001,0.0001]
#step_size = 1.0
x_his = []
iter_his = []
lr_his = []
```

```python
'''
需要執行初始值設定, 下列三個
# Learning rate
#iteration counter
#Step size  
'''
'''
while previous_step_size > precision and iters < max_iters:

  算法迭代更新

print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations
'''
```

```python
for lr in lrs:
    cur_x = 3
    IterNum = 0
    step_size = 1.0
    while step_size > precision and IterNum < max_iters:
        x_his.append(cur_x)
        iter_his.append(IterNum)
        lr_his.append(lr)
        prev_x = cur_x
        cur_x = prev_x - lr*df(prev_x)
        step_size = abs(cur_x - prev_x)
        IterNum += 1
        print("learning rate=",lr,"\tIter=",IterNum,"\tx_value=","%.6f" %(cur_x),"\tstep_size=",step_size)
    x_his.append(cur_x) #last step data
    iter_his.append(IterNum) #last step data
    lr_his.append(lr) #last step data
    if step_size <= precision:
        print("convergence")
    elif IterNum == max_iters:
        print("unconvergence")
```

```python
dataD = {"L_rate":lr_his,
         "Iter":iter_his,
         "X":x_his
        }
LIX_df = pd.DataFrame(dataD)
LIX_df["X"][-1:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for lr in lrs:
    plt.plot(LIX_df[LIX_df["L_rate"]==lr]["Iter"],LIX_df[LIX_df["L_rate"]==lr]["X"]
             ,ms=3,lw=1.5,label="L_rate:%f"%(lr))
    plt.xlabel("ITERs",fontsize=16)
    plt.ylabel("X",fontsize=16)
    plt.title(r'$f(x)=(x+5)^2$',fontsize=16)
    plt.grid()
    plt.legend(loc="upper right",fontsize=16)
plt.show()
```
