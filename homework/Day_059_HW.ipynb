### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

```python
X
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
pca = decomposition.PCA(n_components=3)

pca.fit(X)
X_pca = pca.transform(X)
X_pca
```

```python
fig = plt.figure(1, figsize=(4,3))
ax = Axes3D(fig, rect=[0,0,.95,1], elev=48, azim=134)

for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    ax.text3D(X_pca[y == label,2].mean(),
              X_pca[y == label,0].mean(),
              X_pca[y == label,1].mean()+0.01,name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
ax.scatter(X_pca[:,2],X_pca[:,0],X_pca[:,1],c=y,edgecolor='k')
plt.show()
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_pca2D = pca.transform(X)
X_pca2D
```

```python
plt.scatter(X_pca2D[:,0],X_pca2D[:,1],c=y,edgecolor='k')
for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    plt.text(X_pca2D[y == label,1].mean(),
              X_pca2D[y == label,0].mean(),
              name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
plt.show()
```
