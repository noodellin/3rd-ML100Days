### 用 digits (dataset.load_digits()) 資料嘗試跑 t-SNE

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox
from sklearn import manifold, datasets
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

```python
TSE = manifold.TSNE(n_components=3,random_state=4)
X_TSE = TSE.fit_transform(X)

pca = PCA(n_components=3,random_state=4)
X_PCA = pca.fit_transform(X)
```

```python
fig0 = plt.figure(1, figsize=(4,3))
ax0 = Axes3D(fig0, rect=[0,0,.95,1], elev=48, azim=134)

for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    ax0.text3D(X_TSE[y == label,0].mean(),
              X_TSE[y == label,1].mean(),
              X_TSE[y == label,2].mean()+0.01,name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
ax0.scatter(X_TSE[:,0],X_TSE[:,1],X_TSE[:,2],c=y,edgecolor='k')
plt.title("TSE 3D result",size=18)
plt.xlim(-100, 100)
plt.ylim(-100, 100)



fig1 = plt.figure(2, figsize=(4,3))
ax1 = Axes3D(fig1, rect=[0,0,.95,1], elev=48, azim=134)

for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    ax1.text3D(X_PCA[y == label,0].mean(),
              X_PCA[y == label,1].mean(),
              X_PCA[y == label,2].mean()+0.01,name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
ax1.scatter(X_PCA[:,0],X_PCA[:,1],X_PCA[:,2],c=y,edgecolor='k')
plt.title("PCA 3D result",size=18)
#plt.xlim(-100, 100)
#plt.ylim(-100, 100)

plt.show()
```

```python
TSE_2D = manifold.TSNE(n_components=2,random_state=4)
X_TSE_2D = TSE_2D.fit_transform(X)

pca_2D = PCA(n_components=2,random_state=4)
X_PCA_2D = pca_2D.fit_transform(X)

plt.figure(1, figsize=(4,3))

for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    plt.text(X_TSE_2D[y == label,0].mean(),
              X_TSE_2D[y == label,1].mean(),
              name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
plt.scatter(X_TSE_2D[:,0],X_TSE_2D[:,1],c=y,edgecolor='k')
plt.title("TSE 2D result",size=18)

plt.figure(2, figsize=(4,3))
for name, label in [("label 0", 0),("label 1",1),("label 2",2)]:
    plt.text(X_PCA_2D[y == label,0].mean(),
              X_PCA_2D[y == label,1].mean(),
              name,horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
    
plt.scatter(X_PCA_2D[:,0],X_PCA_2D[:,1],c=y,edgecolor='k')
plt.title("PCA 2D result",size=18)

plt.show()
```
