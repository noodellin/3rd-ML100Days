```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
from pandas import Series, DataFrame
from skimage import io
import requests as req
import codecs
res = req.get("https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt")
f=open("data.txt","w",encoding="UTF-8")
s=res.text
f.write(s)
f.close()
```

```python
with open("data.txt","r",encoding="UTF-8") as f:
    data1=f.readlines()
data2=[]
for i in range(5):
    data2.append(data1[i].strip().split("\t"))
df1=DataFrame(data2)
```

```python
for i in range(len(df1[1])):
    print(df1[1][i])
    image=io.imread(df1[1][i])
    io.imshow(image)
    io.show()
```
