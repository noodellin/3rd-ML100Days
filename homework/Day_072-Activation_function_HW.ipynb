# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

#ReLu function

def ReLu(x):
    if x > 0:
        return(x)
    else:
        return(0)
    
def dReLu(x):
    if x > 0:
        return(1)
    else:
        return(0)
    
x = plt.linspace(-10,10,100)
y=[]
for i in x:
    y.append(ReLu(i))
plt.plot(x,y,"b")
plt.grid()
plt.title("ReLu function",fontsize=15)
plt.text(-2,7,r'$f(x)=\max(0,x)$',fontsize=15)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))
plt.xlabel("X",fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.show()
```

```python
x = plt.linspace(-10,10,200)
y=[]
for i in x:
    y.append(dReLu(i))
plt.plot(x,y,"b", label='linspace(-10,10,10)')
plt.grid()
plt.title("dReLu function",fontsize=15)
plt.text(-8,0.8,r'$f(x)=\max(0,1)$',fontsize=15)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(.1))
plt.xlabel("X",fontsize=15)
plt.ylabel("Y",fontsize=15)
plt.show()
```
